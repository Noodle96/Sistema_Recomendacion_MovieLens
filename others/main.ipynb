```python
# leer en pata la informacion de MovieRatings.csv
data = pd.read_csv("Movie_Ratings.csv", index_col=0)
# print(data)
print(data.columns)
```

```python
# Funciones de distancia
# def euclidean_distance(name1: str, name2: str) -> float:
#     return 1.0
def manhattan_distance(name1: str, name2: str) -> float:
    return 1.0
def pearson_distance(name1: str, name2: str) -> float:
    return 1.0
def cosine_distance(name1: str, name2: str) -> float:
    return 1.0

def euclidean_distance(user1, user2, df):
    # Obtener las calificaciones de ambos usuarios
    ratings1 = df[user1]
    ratings2 = df[user2]

    # Filtrar solo los índices (canciones) donde ambos usuarios tienen calificaciones
    common_mask = ratings1.notna() & ratings2.notna()

    if not common_mask.any():
        return 0.0

    # Calcular la distancia euclidiana
    diff = ratings1[common_mask] - ratings2[common_mask]
    distance = np.sqrt(np.sum(diff ** 2))

    return round(distance, 2)
# dist = euclidean_distance("ANGELICA", "BILL", df)
# print("Distancia:", dist)
```

```python
dist = euclidean_distance("Erin", "Katherine", data)
print(dist)
```

```python
def pearson_correlation(user1, user2, df):
    ratings1 = df[user1]
    ratings2 = df[user2]

    # Solo ítems calificados por ambos
    mask = ratings1.notna() & ratings2.notna()

    if mask.sum() < 2:
        return 0.0  # No se puede calcular Pearson con menos de 2 puntos

    x = ratings1[mask]
    y = ratings2[mask]

    mean_x = x.mean()
    mean_y = y.mean()

    numerator = ((x - mean_x) * (y - mean_y)).sum()
    denominator = np.sqrt(((x - mean_x)**2).sum()) * np.sqrt(((y - mean_y)**2).sum())

    if denominator == 0:
        return 0.0  # No correlación si variancia es cero

    return round(numerator / denominator, 2)

dist = pearson_correlation("matt", "xyx", df)
# print("Distancia:", dist)
```
